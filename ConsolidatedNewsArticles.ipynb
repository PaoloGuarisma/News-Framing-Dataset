```python
import requests
import json
import numpy as np
import pandas as pd
```

### **Get news articles from 3 sources**
#### Source 1 (Space.com)

```python

URL="https://www.space.com/news/archive/2021/02"

page=requests.get(URL)
print(page.content)


from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')

#Getting main contents
news_main = soup.find_all("ul", class_="day-list")

news26 = news_main[2]
news26links = news26.find_all('a', href=True)

news25 = news_main[3]
news25links = news25.find_all('a', href=True)
```

```python
num_articles26= 12

for n in range(1,num_articles26):
    #open link and get contents
    link = news26links[n].get('href')
    article = requests.get(link)
    article_content = article.content
    soup_article = BeautifulSoup(article_content, 'html5lib')
    
    title = soup_article.find('h1').get_text()
    author = soup_article.find('span', style="white-space: nowrap").get_text()
    date = soup_article.find('time', itemprop="datePublished").get_text()
    body = soup_article.find_all(id="article-body")
    w = body[0].find_all('p')
    list_paragraphs = []
    for p in range(0, len(w)-1):
        paragraph = w[p].get_text()
        list_paragraphs.append(paragraph)
        fullarticle = " ".join(list_paragraphs)

    #print the articles
    print('-------------')
    print(f'{title}')
    print(f'  By:{author}  | {date}\n') 
    print(fullarticle)

num_articles25= 9

for m in range(0,num_articles25):
    #open link and get contents
    link = news25links[m].get('href')
    article = requests.get(link)
    article_content = article.content
    soup_article = BeautifulSoup(article_content, 'html5lib')

    title = soup_article.find('h1').get_text()
    author = soup_article.find('span', style="white-space: nowrap").get_text()
    date = soup_article.find('time', itemprop="datePublished").get_text()
    body = soup_article.find_all(id="article-body")
    w = body[0].find_all('p')
    list_paragraphs = []
    for k in range(0, len(w)-1):
        paragraph = w[k].get_text()
        list_paragraphs.append(paragraph)
        fullarticle = " ".join(list_paragraphs)

    #print the articles
    print('-------------')
    print(f'{title}')
    print(f'  By:{author}  | {date}\n') 
    print(fullarticle)
```
